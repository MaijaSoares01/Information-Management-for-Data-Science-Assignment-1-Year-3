### **1. Iterrows** 

**Iterrows()**<br>
DataFrame.iterrows() method/operation allows for us to iterate over DataFrame rows by using index and a series pairing. Iterrows iterates over the data frame column and the operation returns a column name and content from the series. 


***References***<br>
[Python pydata](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iterrows.html)<br>

```python
import pandas as pd
students = pd.DataFrame({'Name': ['Adam', 'Kim', 'Lucy'], 'Age': [9, 10, 20]})
print('Display table:')
print(students)
print('\nDisplay table and call iterrows:')
for index, row in students.iterrows():
    print(row['Name'], row['Age'])
print('\niterrows displays the table without index and column title')
```

### **2. Iloc**

**Iloc()**<br>
Iloc() method/operation allows us to select a certain cell of the dataset. It is used to select a value that belongs to a certain row or column from a dataframe. We are only able to use integer type data values as this represents the index values of the dataframe table for the data values to be accessed and displayed. This operation is extremely useful to look at specific rows or columns of data to analyse and evaluate it.


***References***<br>
[Pandas iloc](https://www.shanelynn.ie/pandas-iloc-loc-select-rows-and-columns-dataframe/)<br>

```python
import pandas as pd
mydict = [{'X': 12, 'Y': 9},
          {'X': 10, 'Y': 10},
          {'X': 19, 'Y': 20},
          {'X': 27, 'Y': 23},
          {'X': 1, 'Y': 13},
          {'X': 7, 'Y': 9},
          {'X': 8, 'Y': 8},
          {'X': 3, 'Y': 6}]
points = pd.DataFrame(mydict)

print('~Use iloc~')
print(type(points.iloc[3:6]))
print(points.iloc[0])
print('\n~Use iloc~')
print(points)
print('\n~Use iloc~')
print('~1~')
print(points.iloc[[0, 1]])
print('\n~2~')
print(points.iloc[2:6])
print('\n~3~')
print(points.iloc[:4])
```

### **3. Drop**

**Drop()**<br>
Drop() method/operation allows for the user to remove or drop certain labels from rows or columns. The operation can remove rows or columns by naming the label and corresponding axis or by naming the index directly or the name of the column. Labels on different levels are also able to be removed by specifying the level the level that they are located, this is used when we have a multi-index table. This is a very beneficial function that is available on Python through Pandas because we are able to modify and correct the data so we can better analyse it.


***References***<br>
[Shanelynn Drop](https://www.shanelynn.ie/pandas-drop-delete-dataframe-rows-columns/)<br>

```python
import pandas as pd
students = pd.DataFrame({'First Name': ['Adam', 'Kim', 'Lucy'],'Last Name': ['Frank', 'Long', 'Art'], 'Age': [9, 10, 20]})
print('Display table:')
print(students)
print('\nDisplay table after dropping Age column:')
print(students.drop(columns=['Age']))
print('\nDisplay table after dropping Row 0 and 1:')
print(students.drop([0,1]))
```

### **5. Apply**

**Apply()**<br>
Apply() method/operation allows a DataFrame table axis to hold a function assigned to it such as a mathematical function. Series objects are Objects that are passed into the function, their index is either the DataFrame’s index (row->axis=0) or the DataFrame’s column(axis=1).


***References***<br>
[W3school Apply](https://www.w3resource.com/pandas/dataframe/dataframe-apply.php)<br>

```python
import pandas as pd
import numpy as np
df = pd.DataFrame([[3, 5]] * 2, columns=['X-axis', 'Y-axis'])
display(df)
display('Display square root of data table:')
display(df.apply(np.sqrt))
print('Display sum of data table (column):')
print(df.apply(np.sum, axis=0))
print('Display sum of data table (row):')
print(df.apply(np.sum, axis=1))
```

### **1. Astype**

**Astype()**<br>
Astype() method/operation allows for a certain column of data type be converted to another type. This is a very useful function as we can change an entire existing columns data type to a more appropriate type quickly and easily. We are able also able to convert a object date type column to a categorical data type column.


***References***<br>
[Astype W3resource](https://www.w3resource.com/pandas/series/series-astype.php)<br>[Astype pydata](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.astype.html)

```python
import pandas as pd
students = pd.DataFrame({'First Name': ['Adam', 'Kim', 'Lucy'],'Last Name': ['Frank', 'Long', 'Art'], 'Age': [9.67, 10, 20]})
before = students
ageBefore = type(students.Age[0])
after = students.astype({"First Name":'category', "Last Name":'category', "Age":'int64'})
ageAfter = type(after.Age[0])
print('Type for Age before astype change:')
print(ageBefore,"\n")
print('Type for Age after astype change:')
print(ageAfter,"\n")
print('Type for Entire table before astype change:')
before.info()
print('\nType for Entire table after astype change:')
after.info()
```

### **2. Drop_duplicates**

**Drop_duplicates()**<br>
Drop_duplicates() method/operation allows for duplicate rows to be removed from the DataFrame table and the table is then returned. This is very useful as being able to drop/remove duplicate values in a table allow for easier data analysis. Data is removed based on all columns by default however you can change this to be based on specific column(s) using subsets. We are also able to keep both the first and last occurrence of a duplicated value. To ensure all duplicate values are dropped/removed we can set keep to False.


***References***<br>
[Python Guide Drop Duplicate](https://pythonguides.com/python-pandas-drop-duplicates/)<br>

```python
import pandas as pd
students = pd.DataFrame({'First Name': ['Adam', 'Kim', 'Lucy', 'Kim', 'Ron', 'Valerie', 'Adam'],'Last Name': ['Frank', 'Long', 'Art', 'Long', 'Frank', 'Crane', 'Frank'], 'Age': [10, 5, 20, 5, 15, 17, 10]})
print("Display Table Students:")
display(students)
print("Display Table Students and drop duplicates:")
display(students.drop_duplicates())
employees = pd.DataFrame({'ID': [16, 89, 13, 55, 89, 97, 11], 'FullName': ['Krish Potts', 'Arnold Wicks', 'Brenna Macfarlane', 'Timothy Randall', 'Arnold Weather', 'Conrad Wilkins', 'Eliot Jackson'],'DepartmentNumber': ['23', '4', '7', '10', '45', '10', '10']})
print("Display Table Employees:")
display(employees)
print("Display Table Employees and drop duplicates:")
display(employees.drop_duplicates(subset=['ID']))
print("Display Table Employees and drop duplicate DepartmentNumbers but keep the Last:")
display(employees.drop_duplicates(subset=['DepartmentNumber'], keep='last'))
```

### **3. Groupby**

**Groupby()**<br>
Groupby() method/operation allows for one or more columns to be grouped together and then we can perform a mathematical operation on them. This is very handy as we can apply mathematical functions and combine the results to better analyse the data. Large amounts of data can be summarized using operations on the groups which can be very useful and is very powerful. Groupby involves splitting the object, summarizing results and applying operations. 


***References***<br>
[Real Python GroupBy ](https://realpython.com/pandas-groupby/)<br>
[Shanelynn grouping](https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/)<br>

```python
import pandas as pd
grades = pd.DataFrame({'StudentName': ['Lucy', 'Megan','Ryan', 'Lucy','Ryan','Lucy', 'Ryan','Megan','Megan'],'Grade': [45,98,70,30,75,48,80,90,95]})
print("Display Table Grades:")
display(grades)
print("Display Table Grades using Groupby by StudentName:")
display(grades.groupby(['StudentName']).mean())
print("Display Table Grades Groupby by StudentName and summed up grades:")
display(grades.groupby(by=["StudentName"]).sum())
numbers = [[8, 9, 5], [None, 10, None], [2, 1, 5], [3, None, 6]]
list = pd.DataFrame(numbers, columns=["x", "y", "z"])
print("Display Table List:")
display(list)
print("Display Table List Groupby by Y and summed up values:")
list.groupby(by=["y"], dropna=False).sum()
```

### **4. Merge**

**Merge()**<br>
Merge() method/operation allows for data to be combined on common columns or indices. This is very handy as we can combine data objects based on one or many columns in a similar way. It is very handy as we can merge rows that share data. We can display and merge the data using joins logic e.g., inner, right, left, and cross.

***References***<br>
[Shanelynn Merge](https://www.shanelynn.ie/merge-join-dataframes-python-pandas-index-1/)<br>
[Real Python Merge](https://realpython.com/pandas-merge-join-and-concat/)<br>

```python
import pandas as pd
list1 = pd.DataFrame({'Name': ['Mark', 'Megan', 'Tanya', 'Sasha','Abby'],
                    'Grade': [45, 60, 70, 90,88]})
list2 = pd.DataFrame({'StudentName': ['Sasha', 'Tanya', 'Megan', 'Mark','Ryan'],
                    'Grade': [99, 75, 63, 47,30]})
print("Display Table list1:")
display(list1)
print("Display Table list2:")
display(list2)
newList1 = list1.merge(list2, left_on='Name', right_on='StudentName')
print("Display Merged Table List:")
display(newList1)
newList2 = list1.merge(list2, left_on='Name', right_on='StudentName', suffixes=('_left', '_right'))
print("Display Merged Table List with slight change to column names:")
display(newList2)
l1 = pd.DataFrame({'Name': ['Mark', 'Megan', 'Tanya', 'Sasha','Abby'], 'Test': [45, 60, 70, 90,88]})
l2 = pd.DataFrame({'Name': ['Sasha', 'Tanya', 'Megan', 'Mark','Ryan'], 'Grade': [99, 75, 63, 47,30]})
newList3 = l1.merge(l2, how='inner', on='Name')
print("Display Merged Table List merged on inner, based on Name(similar column):")
display(newList3)
newList4 = l1.merge(l2, how='left', on='Name')
print("Display Merged Table List merged on left, based on Name(similar column):")
display(newList4)
newList5 = l1.merge(l2, how='right', on='Name')
print("Display Merged Table List merged on right, based on Name(similar column):")
display(newList5)
```
